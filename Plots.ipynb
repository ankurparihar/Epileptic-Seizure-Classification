```python
def plot_graph(x_data, y_data, title):
    plt.plot(x_data, y_data)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.show()
def plot_graph_lim(x_data, y_data, lim_l, lim_h, title):
    plt.plot(x_data, y_data)
    plt.plot(x_data, [lim_l] * len(x_data), x_data, [lim_h] * len(x_data), color='lime')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.show()
def plot_graph_2(x1_data, y1_data, x2_data, y2_data, title):
    plt.plot(x2_data, y2_data, x1_data, y1_data)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.show()
def plot_graph_lim_2(x_data, y_data, part, lim_l, lim_h, title):
    x1_data = x[:part]
    x2_data = x[part:]
    y1_data = y[:part]
    y2_data = y[part:]
    plt.plot(x2_data, y2_data, x1_data, y1_data)
    plt.plot(x1_data, [lim_l] * len(x1_data), x2_data, [lim_l] * len(x2_data), x1_data, [lim_h] * len(x1_data), x2_data, [lim_h] * len(x2_data), color='lime')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.show()
```

```python
# ---------------- Shifted waves ---------------- #
```

```python
x = list(range(1800))
Y = []
```

```python
y = [math.sin(math.pi*p/180) for p in x]
plot_graph(x,y,'y=sin(x)')
Y.append(y)
```

```python
y = [math.cos(math.pi*p/180) for p in x]
plot_graph(x,y,'y=cos(x)')
Y.append(y)
```

```python
y = [max(-10,min(10,math.tan(math.pi*p/180))) for p in x]
plot_graph(x,y,'y=tan(x)')
Y.append(y)
```

```python
model = pulseNet(pos_lim=0.8, neg_lim=-0.8, pulse_len=4)
model.transform_pulse(Y)
matrix = np.asarray(model.pulses_to_matrix(), dtype=np.ndarray)
```

```python
model.plot_matrix(matrix[0])
model.plot_matrix(matrix[1])
model.plot_matrix(matrix[2])
```

```python
# ---------------- Wave order ---------------- #
```

```python
x = list(range(1800))
Y = []
```

```python
y1 = [math.sin(math.pi*p/180) for p in x]
plot_graph(x,y1,'y=sin(x)')
y2 = [max(-10,min(10,math.tan(math.pi*p/180))) for p in x]
plot_graph(x,y2,'y=tan(x)')
```

```python
y3 = y1 + y2
y4 = y2 + y1
Y.append(y3)
Y.append(y4)
plot_graph(x+[xi+1800 for xi in x],y3,'y=sin_tan(x)')
plot_graph(x+[xi+1800 for xi in x],y4,'y=tan_sin(x)')
```

```python
model = pulseNet(pos_lim=0.5, neg_lim=-0.5, pulse_len=4)
model.transform_pulse(Y)
matrix = np.asarray(model.pulses_to_matrix(), dtype=np.ndarray)
model.plot_matrix(matrix[0])
model.plot_matrix(matrix[1])
```
